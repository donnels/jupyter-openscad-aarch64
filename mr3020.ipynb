## initial measurements
- 6.7cm wide
- 7.4cm deep
- 2.2cm high
rounded corners
ethernet port
usb port and micro usb port for power
swith on side and leds and button on top

```openscad
// Router
height=22;
//width=67;
width=67;
//depth=74;
depth=74;
#cube([width,depth,height]);
```

## the initial Cube without rounded corners

```openscad
%display
```

## adding the subtraction points for the rounded corners

```openscad
cubeSide=6;
cubeSize=[cubeSide+1,cubeSide+1,height+2];
cube(cubeSize);
translate ([width-cubeSide,0,-1]) cube(cubeSize);
translate ([width-cubeSide,depth-cubeSide,-1]) cube(cubeSize);
translate ([0,depth-cubeSide,-1]) cube(cubeSize);
```

```openscad
%display
```

## adding the top LEDs

```openscad
//LEDs on top
ledsOffset=[8,14,-1];
ledsSize=[2,40,2];
translate (ledsOffset+[0,0,height]) color([0,1,0]) cube(ledsSize);
```

```openscad
%display
```

## Adding the ethernet port

The ethernet port is shown extruded so as to be easily subtracted from the holder.

```openscad
//ethernet port on side
eth3D=40;
ethSize=[19.5,eth3D,13];
ethOffset=[21,-(eth3D/2),6.7];
translate (ethOffset+[0,0,0]) color([0,1,0]) cube(ethSize);
```

```openscad
%display
```

## Adding the Power Port
The Power Port is shown extruded to be easily subtracted from the main body.

```openscad
//power connexion on side
pwr3D=40;
pwrSize=[12.3,pwr3D,5];
pwrOffset=[42,-(pwr3D/2),6];
translate (pwrOffset+[0,0,0]) color([1,0,0]) cube(pwrSize);
```

```openscad
%display
```

## Adding the switch on the side of Power- and LAN ports
The switch is extruded to be easily substracted from the main body.


```openscad
//Mode switch
swt3D=40;
swtSize=[7.1,swt3D,3.6];
swtOffset=[12,-(swt3D/2),5.2];
translate (swtOffset+[0,0,0]) color([0,0,1]) cube(swtSize);
```

```openscad
%display
```

## Adding the USB Port on it's  own side
The USB Port is extruded to be easily substracted from the main body.

```openscad
//USB Port
USB3D=40;
USBSize=[USB3D,14.5,6.9];
USBOffset=[width-(USB3D/2),11,8.5];
translate (USBOffset) color([1,1,1]) cube(USBSize);
```

```openscad
%display
```
