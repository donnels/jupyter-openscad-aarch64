## Initialize variables etc

```openscad
$fn=360;
SPlength=170;
SPwidth=28;
SPmaterialStrength=2;
SPDepth=6;
offset=-SPDepth;
version="Strikeplate V5 2022-01-01";
//for versioning
module letter(string) {
    textFont = "Liberation Sans";
    letter_size = 5;
    letter_height = 2;
    linear_extrude(height = letter_height) {
        text(string, size = letter_size, font = textFont, halign = "center", valign = "center");
    }
};
```

```openscad
%%display letter(version);
```

## First primitives for testing

```openscad
;
module strikePlate () {
    cube ([SPlength,SPwidth,SPmaterialStrength]);
    translate ([0,0,offset])
        cube ([SPlength,SPmaterialStrength,SPDepth]);
    translate ([SPlength/2,SPwidth/2,0]) 
        rotate([0,180,0]) 
            translate([0,0,-1]) 
                letter(version);
}
;
```

```openscad
%%display strikePlate();
```

## Tool(s) to shape rounded corners by subtraction
The following modules are kept simple and could easily be pulled together as one with some options.
These can be considered in a prototype stage.

```openscad

;
shaperRad=9;
shaperDepth=2.11;
shaperHeight=4;
module cornerShaperL () {
        difference() {
            translate([-.1,-.1,-.1]) 
                cube([shaperRad+.1,shaperDepth,shaperRad+.1]);
            translate([shaperRad,shaperHeight-.1,shaperRad])
                rotate([90,0,0]) 
                    cylinder(h=shaperHeight+.2,r=shaperRad);
        }
}
module cornerShaperR () {
        difference() {
            translate([0,-.1,-.1]) 
                cube([shaperRad+.1,shaperDepth,shaperRad+.1]);
            translate([0,shaperHeight,shaperRad])
                rotate([90,0,0]) 
                    cylinder(h=shaperHeight+.2,r=shaperRad);
        }
}
edgeShaperRad=2;
edgeShaperLength=70;
module edgeShaper (edgeShaperLength,edgeShaperRad) {
        difference() {
            translate([0,-.1,-.1]) 
                cube([edgeShaperLength,edgeShaperRad+.2,edgeShaperRad+.2]);
            translate ([-.1,edgeShaperRad,0])
                rotate([0,90,0])
                    cylinder(h=edgeShaperLength+.2,r=edgeShaperRad);
            }
}


%%display cornerShaperL();#cube([shaperRad+.1,shaperRad+.1,shaperRad+.1]);
```

```openscad
%%display edgeShaper(10,2);
```

```openscad
%%display 
difference() {
    strikePlate();
    #translate([-1,0,offset]) cornerShaperL();
    #translate([SPlength-shaperRad+1,0,offset]) cornerShaperR();
    #translate([-.1,0,0]) edgeShaper(SPlength+.2,2);
}
```

## Screws for subtraction

```openscad
;
module screw () {
    cylinder(h=3,d1=2,d2=7);
}
;
```

```openscad
%%display screw();
```

```openscad
;
difference(){
    strikePlate();
    translate([8.5,10.5,-0.1]) screw();
    translate([85,10.5,-0.1]) screw();
    translate([SPlength-8.5,10.5,-0.1]) screw();
    translate([-1,0,offset]) cornerShaperL();
    translate([SPlength-shaperRad+1,0,offset]) cornerShaperR();
    translate([-.1,0,0]) edgeShaper(SPlength+.2,2);
}
;
```
