```openscad
%clear
```

## initialize variables etc

```openscad
$fn=100;
SPlength=170;
SPwidth=28;
SPmaterialStrength=2;
offset=-10;
;
```

## First primitives for testing

```openscad
;
module strikePlate () {
    color ([1,1,1]) {
        cube ([SPlength,SPwidth,SPmaterialStrength]);
        translate ([0,0,offset])
            cube ([SPlength,SPmaterialStrength,10]);
    }
}
;
```

```openscad
%%display strikePlate();
```

## Tool to shape rounded corners by subtraction

```openscad

;
$fn=360;
cornerRad=9;
cornerHeight=10;
module cornerShaperL () {
        difference() {
            translate([-.1,-.1,-.1]) 
                cube([cornerRad+.1,cornerRad+.1,cornerRad+.1]);
            translate([cornerRad,cornerHeight-.1,cornerRad])
                rotate([90,0,0]) 
                    cylinder(h=cornerHeight+.2,r=cornerRad);
        }
}
module cornerShaperR () {
        difference() {
            translate([0,0,-.1]) 
                cube([cornerRad+.1,cornerRad+.1,cornerRad+.1]);
            translate([0,cornerHeight,cornerRad])
                rotate([90,0,0]) 
                    cylinder(h=cornerHeight+.2,r=cornerRad);
        }
}

%%display cornerShaperL();#cube([cornerRad+.1,cornerRad+.1,cornerHeight]);
```

```openscad
%%display 
difference() {
    strikePlate();
    translate([0,0,offset]) cornerShaperL();
    translate([SPlength-cornerRad,-.1,offset]) cornerShaperR();
}
```

## Screws for subtraction

```openscad
;
module screw () {
    cylinder(h=3,d1=2,d2=7);
}
;
```

```openscad
%%display screw();
```

```openscad
;
difference(){
    strikePlate();
    translate([8.5,10.5,-0.1]) screw();
    translate([85,10.5,-0.1]) screw();
    translate([SPlength-8.5,10.5,-0.1]) screw();
    translate([0,0,offset]) cornerShaperL();
    translate([SPlength-cornerRad,-.1,offset]) cornerShaperR();

}
;
```
