# A dummy strike plate for the now non existent lock due to the sliding glass door

```openscad
%clear
```

## initialize variables etc

```openscad
$fn=100;
SPlength=170;
SPwidth=28;
SPmaterialStrength=2;
;
```

```openscad
%command xvfb-run openscad
```

## First primitives for testing

```openscad
;
module strikePlate () {
    color ([1,1,1]) {
        cube ([SPlength,SPwidth,SPmaterialStrength]);
        translate ([0,0,-10])
            cube ([SPlength,SPmaterialStrength,10]);
    }
}
;
```

```openscad
%%display strikePlate();
```

## Screws for subtraction

```openscad
;
module screw () {
    cylinder(h=3,d1=2,d2=7);
}
;
```

```openscad
%%display screw();
```

```openscad
;
difference(){
    strikePlate();
    translate([8.5,10.5,-0.1]) screw();
    translate([85,10.5,-0.1]) screw();
    translate([SPlength-8.5,10.5,-0.1]) screw();
}
;
```
